```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/content/dataset.csv'
df = pd.read_csv('/content/dataset.csv')
df
```

```python
# Membuat Heatmap
plt.figure(figsize=(10, 6))
heatmap_df = df.pivot('Kabupaten', 'Tahun', 'Jumlah')
sns.heatmap(heatmap_df, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Populasi Kabupaten Sulawesi Tenggara (2014-2023)')
plt.xlabel('Tahun')
plt.ylabel('Kabupaten')
plt.tight_layout()
plt.show()
```

Visualisasi Heatmap di atas menunjukkan bahwa populasi di Sulawesi Tenggara meningkat secara keseluruhan dalam beberapa tahun terakhir. Kota Kendari menjadi daerah dengan populasi terbanyak, sedangkan Konawe Kepulauan menjadi daerah dengan populasi terendah.
Sementara itu, Kabupaten Konawe menjadi daerah dengan populasi yang pesat. Sedangkan Kabupaten Muna menjadi daerah dengan populasi yang relatif stabil.
